```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def iprsatufasa():
    Qo = float(input("Masukkan nilai laju alir minyak (bopd) : "))
    Press = float(input("Masukkan nilai Tekanan Reservoir (Pr) atau Tekanan Statik Sumur (Ps) (psi) : "))
    Pwf = float(input("Masukkan nilai Tekanan Bottomhole (Pwf) : "))
    PI = Qo / (Press - Pwf)
    print("Nilai Productivity Index (PI) adalah", str(PI), "bopd/psi.")
    Qomax = PI * Press
    print("Nilai laju alir minyak maksimum (Qomax) adalah", str(Qomax), "bopd")
    Pwf = 0
    daftaripr = []
    while Pwf <= Press:
        Qo = PI * (Press - Pwf)
        daftaripr.append({"Bottomhole Pressure, Pwf (psi)": str(Pwf), "Laju Alir Produksi Minyak, Qo (bopd)": str(Qo)})
        Pwf += Press/10
    df = pd.DataFrame(daftaripr)
    df["Bottomhole Pressure, Pwf (psi)"] = pd.to_numeric(df["Bottomhole Pressure, Pwf (psi)"])
    df["Laju Alir Produksi Minyak, Qo (bopd)"] = pd.to_numeric(df["Laju Alir Produksi Minyak, Qo (bopd)"])
    plt.figure(figsize=(10,6))
    plt.plot(df["Bottomhole Pressure, Pwf (psi)"], df["Laju Alir Produksi Minyak, Qo (bopd)"])
    plt.xlabel("Bottomhole Pressure, Pwf (psi)")
    plt.ylabel("Laju Alir Produksi Minyak, Qo (bopd)")
    plt.title("Kurva IPR Present Metode Standing")
    plt.grid(True)
    plt.show
    return df
```

```python
iprsatufasa()
```
